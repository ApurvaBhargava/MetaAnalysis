## For 7 sets: orkney, gabon, tanzania, madagascar/kenya, vietnam, cambodia, madagascar

```python
df7 = pd.read_csv('combined7.csv', index_col=0)
print("Size of dataset: ", len(df7))
print("Number of null target outputs: ", df7['envProduction'].isna().sum())
print("Usable size of dataset: ", len(df7)-df7['envProduction'].isna().sum())
```

```python
print("Null/NaN values in every attribute for every dataset:")

col_list = sorted(df7.columns)
loss_matrix = df7.groupby('country')[col_list[0]].apply(lambda x: x.isna().sum())
for col in col_list[1:]:
    loss_matrix = pd.concat([loss_matrix, df7.groupby('country')[col].apply(lambda x: x.isna().sum())], axis=1)
loss_matrix.T.to_csv('df7_loss.csv')
loss_matrix.T
```
